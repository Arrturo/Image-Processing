LINES DETECTION

```python
photo = Image(
    "D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lines.jpg")
result_photo = Image(photo.data, color_model=ColorModel.RGB)
photo.show_img()
```

```python
gray = photo.to_gray()
thresh = gray.otsu()

thresh.show_img()
```

```python
edges = thresh.canny(20, 50, 3)
edges.show_img()
```

```python
lines = edges.hough_lines(result_photo, (0, 255, 0))
result_photo.show_img()
```
