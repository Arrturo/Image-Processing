```python
import sys

sys.path.append('../')
from Image import *
import numpy as np


photo = Image('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo1 = Image('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo2 = Image('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
```

Types of kernels:

```python

# Filtr tożsamościowy
identity = np.array( [[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]] )

# Filtr górnoprzepustowy
sharpen = np.array( [[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]] )

# Filtr dolnoprzepustowy (normalized)
box_blur = np.array( [[1, 1, 1],
                    [1, 1, 1],
                    [1, 1, 1]] )

# Rozmycie Gaussowskie 3x3
gaussian_blur = np.array(  [[1, 2, 1],
                            [2, 4, 2],
                            [1, 2, 1]] )

# Rozmycie Gaussowskie 5x5
gaussian_blur2 = np.array( [[1, 4, 6, 4, 1],
                            [4, 16, 24, 16, 4],
                            [6, 24, 36, 24, 6],
                            [4, 16, 24, 16, 4],
                            [1, 4, 6, 4, 1]] )

#Operator Sobela
w0 = np.array( [[-1, 0, 1],
                [-2, 0, 2],
                [-1, 0, 1]] )

w45 = np.array( [[0, 1, 2],
                [-1, 0, 1],
                [-2, -1, 0]] )

w90 = np.array( [[1, 2, 1],
                [0, 0, 0],
                [-1, -2, -1]] )

w135 = np.array( [[2, 1, 0],
                [1, 0, -1],
                [0, -1, -2]] )
```

```python
gray = photo.to_gray()
Image.conv_2d(gray, identity)

gray1 = photo1.to_gray()
Image.conv_2d(gray1, box_blur)

gray2 = photo2.to_gray()
Image.conv_2d(gray2, sharpen)
```

```python
Image.show_images(gray, gray1, gray2)
```
