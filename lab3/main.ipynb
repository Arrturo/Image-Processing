```python
from GrayScaleTransform import *
import matplotlib.pyplot as plt


photo = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo1 = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo2 = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo3 = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo4 = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo5 = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
photo6 = GrayScaleTransform('D://Dokumenty//Code//Python//Introduction-to-Computer-Graphics//data//lena.jpg')
```

```python
sepia = photo1.to_sepia(alpha_beta=(1.1, 0.9))
sepia1 = photo2.to_sepia(alpha_beta=(1.5, 0.5))
sepia2 = photo3.to_sepia(alpha_beta=(1.9, 0.1))
```
